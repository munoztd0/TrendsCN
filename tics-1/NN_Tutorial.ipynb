```python
import torch
```

```python
xtrain=torch.rand(150,1)
print(xtrain)
xtest=torch.rand(500,1)
print(xtest)
xval=torch.rand(500,1)
print(xval)
```

```python
def trueModel(x):
  return 2*torch.sin(14*x)+torch.rand(x.size(0),1)
```

```python
ytrain=trueModel(xtrain)
ytest=trueModel(xtest)
yval=trueModel(xval)
```

```python
import torch.nn as nn
netLinear = nn.Sequential(nn.Linear(1,1))

netMLP1 = nn.Sequential(nn.Linear(1,1),nn.ReLU(), nn.Linear(1, 1))
netMLP2 = nn.Sequential(nn.Linear(1,2),nn.ReLU(), nn.Linear(2, 1))
netMLP4 = nn.Sequential(nn.Linear(1,4),nn.ReLU(), nn.Linear(4, 1))
netMLP8 = nn.Sequential(nn.Linear(1,8),nn.ReLU(), nn.Linear(8, 1))
netMLP16 = nn.Sequential(nn.Linear(1,16),nn.ReLU(), nn.Linear(16, 1))
netMLP32 = nn.Sequential(nn.Linear(1,32),nn.ReLU(), nn.Linear(32, 1))
netMLP64 = nn.Sequential(nn.Linear(1,64),nn.ReLU(), nn.Linear(64, 1))
netMLP128 = nn.Sequential(nn.Linear(1,128),nn.ReLU(), nn.Linear(128, 1))
netMLP256 = nn.Sequential(nn.Linear(1,256),nn.ReLU(), nn.Linear(256, 1))
netMLP512 = nn.Sequential(nn.Linear(1,512),nn.ReLU(), nn.Linear(512, 1))
netMLP1024 = nn.Sequential(nn.Linear(1,1024),nn.ReLU(), nn.Linear(1024, 1))
```

```python
net=netMLP16
```

```python
import torch.optim as optim
criterion = nn.MSELoss()
optimizer = optim.Adam(net.parameters(), lr=0.001)
```

```python
for i in range (0,5000) :
  optimizer.zero_grad()
  yPred=net(xtrain)
  error=criterion(yPred,ytrain)
  error.backward()
  optimizer.step()
```
